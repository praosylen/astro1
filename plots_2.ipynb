```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
with fits.open("../Virgo/tables/nsa_mstar.virgo.fits") as tab:
    data = tab[1].data
    cols = tab[1].columns

rvel = ((data["Z"]+1)**2-1)/((data["Z"]+1)**2+1)*3e5 #Recessional velocity in km/s, converted from redshift


#Colors in order of increasing Mstar/decreasing recessional velocity:
#  red --> orange --> yellow --> green --> aqua --> blue --> purple
fig1 = plt.figure(1)
plt.scatter(data["RA"], data["DEC"], c=data["MSTAR_50"], marker=".", s=1, cmap="gist_rainbow")
plt.xlabel("RA")
fig1.axes[0].set_xlim(left=max(data["RA"]), right=min(data["RA"]))
plt.ylabel("DEC")
plt.colorbar(label="Mstar")
plt.title("Sky position vs. Mstar")

fig2 = plt.figure(2)
plt.scatter(data["RA"], data["DEC"], c=rvel, marker=".", s=1, cmap="gist_rainbow_r")
plt.xlabel("RA")
fig2.axes[0].set_xlim(left=max(data["RA"]), right=min(data["RA"]))
plt.ylabel("DEC")
plt.colorbar(label="Recessional velocity (km/s)")
plt.title("Sky position vs. recessional velocity")
pass
```
